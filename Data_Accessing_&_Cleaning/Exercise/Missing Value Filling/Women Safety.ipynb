<a href="#table-of-contents" style="font-family: Arial, sans-serif; font-size: 15px; background-color: #7b2cbf; color: #ffffff; padding: 7px 15px; text-decoration: none; border-radius: 15px;">

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Assuming df is your original DataFrame, create a copy for df10
df10 = dfc.copy()

# Separate rows with and without NaN in the target column
train_data = df10[df10['attitudes_toward_violence_2019'].notna()]
test_data = df10[df10['attitudes_toward_violence_2019'].isna()]

# Identify categorical columns to encode (e.g., 'country')
categorical_cols = ['country'] 

# Apply one-hot encoding to categorical columns in both train and test data
train_data_encoded = pd.get_dummies(train_data, columns=categorical_cols, drop_first=True)
test_data_encoded = pd.get_dummies(test_data, columns=categorical_cols, drop_first=True)

# Align train and test data columns (in case they have different categories)
train_X = train_data_encoded.drop(columns=['attitudes_toward_violence_2019'])
train_y = train_data_encoded['attitudes_toward_violence_2019']

# If the test data has different columns after encoding, align with train
test_X = test_data_encoded.reindex(columns=train_X.columns, fill_value=0)

# Fill NaN values in test_X if they exist
if test_X.isnull().values.any():
    test_X.fillna(0, inplace=True)  # Fill NaNs with 0 or choose another strategy

# Fit the model
model = LinearRegression()
model.fit(train_X, train_y)

# Predict missing values in the test_data and fill them in df10
df10.loc[df10['attitudes_toward_violence_2019'].isna(), 'attitudes_toward_violence_2019'] = model.predict(test_X)

# Optionally, check the updated DataFrame
print(df10[['attitudes_toward_violence_2019']])
```
